```python
import pandas as pd
import numpy as np
tags = pd.read_excel('MT.xlsx')
tags.to_csv ('tagsData.csv', index = False)
```

```python
tags.head(10)
```

```python
tags.drop(tags.filter(regex="Unnamed: 0"),axis=1, inplace=True)
tags.head(10)
```

```python
tags1 = tags.transpose()
tags1 = tags1.rename(index = {'movieId' : 'Tags'})
headers = tags1.iloc[0]
tags1  = pd.DataFrame(tags1.values[1:], columns = headers)
tags1
```

```python
import sklearn
import matplotlib.pyplot as plt
import sklearn.cluster as cluster
from sklearn.cluster import KMeans
```

```python
make_array = tags1.values
make_array
```

```python
K = range(1, 300)
wss = []
for k in K:
    kmeans = KMeans(init = 'k-means++', n_clusters = k, n_init = 10).fit(make_array)
    kmeans = kmeans.fit(make_array)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python
mycenters = pd.DataFrame({'Clusters' : K, 'WSS' : wss})
mycenters
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x = 'Clusters', y = 'WSS', data = mycenters, marker = '+')
```

```python
km = KMeans(n_clusters = 300)
km
```

```python
y_predicted = km.fit_predict(make_array)
y_predicted
tags1['Cluster'] = y_predicted
tags1.set_index('Cluster', inplace = True)
tags1.head(20)
```
